```python
# Import pandas library
import pandas as pd

# Load the CSV file into a DataFrame (adjust path if file is not in the current directory)
df = pd.read_csv('D:\SHE CARES\Menstural_cyclelength.csv')

# Display the first 5 rows of the dataset
df.head()
```

```python
import pandas as pd

# Load the Excel file
df = pd.read_excel('D:/SHE CARES/PCOS_data_without_infertility.xlsx')

# Print missing values by column before filling
print("Missing values before filling:")
print(df.isnull().sum())

# Fill missing numeric values with median of each column
df['Unnamed: 2'].fillna('Unknown', inplace=True)


# Fill missing non-numeric (object) columns optionally, e.g., with mode
for col in df.select_dtypes(include=['object']).columns:
    mode = df[col].mode()
    if not mode.empty:
        df.loc[:, col] = df[col].fillna(mode[0])


# Print missing values after filling
print("\nMissing values after filling:")
print(df.isnull().sum())

# Save the cleaned data back to Excel
df.to_excel('D:/SHE CARES/PCOS_data_without_infertility.xlsx', index=False)
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('D:/SHE CARES/post natal data.csv')

# Fill missing numeric columns with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill missing categorical columns with mode
for col in df.select_dtypes(include=['object']).columns:
    mode = df[col].mode()
    if not mode.empty:
        df.loc[:, col] = df[col].fillna(mode[0])

# Verify no missing values remain
print(df.isnull().sum())

# Save the cleaned data back to Excel
df.to_excel('D:/SHE CARES/PCOS_data_without_infertility.xlsx', index=False)

```
