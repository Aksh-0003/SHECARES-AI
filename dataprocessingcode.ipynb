```python
#PCOS_data_without_infertility.xlsx
import pandas as pd

from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Load Excel file
pcos_df = pd.read_excel('D:/SHE CARES/PCOS_data_without_infertility.xlsx')

# Fill missing numeric columns with median
pcos_df.fillna(pcos_df.median(numeric_only=True), inplace=True)

# Fill missing categorical columns with mode, with safe assignment to avoid warnings
for col in pcos_df.select_dtypes(include=['object']).columns:
    mode = pcos_df[col].mode()
    if not mode.empty:
        pcos_df.loc[:, col] = pcos_df[col].fillna(mode[0])

# Encode categorical columns using Label Encoding (adjust columns as needed)
label_enc_cols = ['Timestamp', 'Age','Feeling sad or Tearful','Irritable towards baby & partner','Trouble sleeping at night','Problems concentrating or making decision','Overeating or loss of appetite','Feeling anxious','Feeling of guilt','Problems of bonding with baby','Suicide attempt']  # replace with actual column names
le = LabelEncoder()
for col in label_enc_cols:
    pcos_df[col] = le.fit_transform(pcos_df[col])

# Normalize numerical features between 0 and 1
num_cols = pcos_df.select_dtypes(include=['float64', 'int64']).columns
scaler = MinMaxScaler()
pcos_df[num_cols] = scaler.fit_transform(pcos_df[num_cols])

# pcos_df is now preprocessed and ready for model training
```

```python
#PCOS_data.csv
import pandas as pd

from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Load Excel file
pcos_df = pd.read_csv('D:/SHE CARES/PCOS_data.csv')

# Fill missing numeric columns with median
pcos_df.fillna(pcos_df.median(numeric_only=True), inplace=True)

# Fill missing categorical columns with mode, with safe assignment to avoid warnings
for col in pcos_df.select_dtypes(include=['object']).columns:
    mode = pcos_df[col].mode()
    if not mode.empty:
        pcos_df.loc[:, col] = pcos_df[col].fillna(mode[0])

# Encode categorical columns using Label Encoding (adjust columns as needed)
label_enc_cols = ['Sl. No','Patient File No.','PCOS (Y/N)',' Age (yrs)','Weight (Kg)','Height(Cm) ','BMI','Blood Group','Pulse rate(bpm) ','RR (breaths/min)'
                  ,'Hb(g/dl)'	,'Cycle(R/I)',	'Cycle length(days)','Marraige Status (Yrs)','Pregnant(Y/N)','No. of abortions','  I   beta-HCG(mIU/mL)',
                  'II    beta-HCG(mIU/mL)','FSH(mIU/mL)'	,'LH(mIU/mL)'	,'FSH/LH','Hip(inch)'	,'Waist(inch)',	'Waist:Hip Ratio',
                  'TSH (mIU/L)','AMH(ng/mL)','PRL(ng/mL)','Vit D3 (ng/mL)','PRG(ng/mL)','RBS(mg/dl)',
                  'Weight gain(Y/N)','hair growth(Y/N)',	'Skin darkening (Y/N)','Hair loss(Y/N)','Pimples(Y/N)'
                  ,'Fast food (Y/N)'	,'Reg.Exercise(Y/N)','BP _Systolic (mmHg)','BP _Diastolic (mmHg)','Follicle No. (L)','Follicle No. (R)',
                  'Avg. F size (L) (mm)','Avg. F size (R) (mm)','Endometrium (mm)']


le = LabelEncoder()
for col in label_enc_cols:
    pcos_df[col] = le.fit_transform(pcos_df[col])

# Normalize numerical features between 0 and 1
num_cols = pcos_df.select_dtypes(include=['float64', 'int64']).columns
scaler = MinMaxScaler()
pcos_df[num_cols] = scaler.fit_transform(pcos_df[num_cols])
print("done processing..")

# pcos_df is now preprocessed and ready for model training
```
