```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

df = pd.read_csv('D:/SHE CARES/PCOS_data.csv')

target_col = 'PCOS (Y/N)'
drop_cols = ['Sl. No', 'Patient File No.']

le = LabelEncoder()
df[target_col] = le.fit_transform(df[target_col].astype(str).str.strip())

X = df.drop(columns=drop_cols + [target_col])
y = df[target_col]

for col in X.select_dtypes(include='object').columns:
    X[col] = LabelEncoder().fit_transform(X[col].astype(str).str.strip())

X = X.fillna(X.median())

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Evaluation for PCOS_data.csv")
print(classification_report(y_test, y_pred, target_names=[str(i) for i in le.classes_]))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load dataset
df = pd.read_csv('D:/SHE CARES/FedCycleData071012.csv')

# Define target and drop columns
target_col = 'TotalDaysofFertility'  # Adjust if needed
drop_cols = ['ClientID', 'Wedding', 'Methoddate', 'Religion', 'Ethnicity']

# Filter rare classes that have more than 1 instance
counts = df[target_col].value_counts()
valid_classes = counts[counts > 1].index
df_filtered = df[df[target_col].isin(valid_classes)].copy()

# Encode target variable
le = LabelEncoder()
df_filtered[target_col] = le.fit_transform(df_filtered[target_col].astype(str).str.strip())

# Prepare features and target
X = df_filtered.drop(columns=drop_cols + [target_col])
y = df_filtered[target_col]

# Encode categorical features
for col in X.select_dtypes(include='object').columns:
    X[col] = LabelEncoder().fit_transform(X[col].astype(str).str.strip())

# Fill missing values
X = X.fillna(X.median())

# Stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# Train Random Forest model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Filter labels for classification report
unique_labels = np.unique(np.concatenate([y_test, y_pred]))
filtered_target_names = [str(le.classes_[label]) for label in unique_labels]

# Print classification report with zero_division=0 to suppress warnings
print("Classification report:")
print(classification_report(
    y_test, y_pred,
    labels=unique_labels,
    target_names=filtered_target_names,
    zero_division=0
))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Load dataset
df = pd.read_csv('D:/SHE CARES/FedCycleData071012.csv')

# Define target and drop columns
target_col = 'TotalDaysofFertility'  # Adjust if needed
drop_cols = ['ClientID', 'Wedding', 'Methoddate', 'Religion', 'Ethnicity']

# Filter rare classes
counts = df[target_col].value_counts()
valid_classes = counts[counts > 1].index
df_filtered = df[df[target_col].isin(valid_classes)].copy()

# Encode target variable
le = LabelEncoder()
df_filtered[target_col] = le.fit_transform(df_filtered[target_col].astype(str).str.strip())

# Prepare features and target
X = df_filtered.drop(columns=drop_cols + [target_col])
y = df_filtered[target_col]

# Encode categorical features
for col in X.select_dtypes(include='object').columns:
    X[col] = LabelEncoder().fit_transform(X[col].astype(str).str.strip())

X = X.fillna(X.median())

# Stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# Define parameter grid for RandomizedSearchCV
param_dist = {
    'n_estimators': [100, 200, 500],
    'max_depth': [None, 10, 20, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Initialize Random Forest and RandomizedSearchCV
rf = RandomForestClassifier(random_state=42)
random_search = RandomizedSearchCV(
    rf, param_distributions=param_dist, n_iter=20, cv=5,
    scoring='accuracy', random_state=42, n_jobs=-1
)

# Fit RandomizedSearchCV
random_search.fit(X_train, y_train)

print("Best hyperparameters:", random_search.best_params_)
print("Best CV accuracy:", random_search.best_score_)

best_model = random_search.best_estimator_

# Predict on test data
y_pred = best_model.predict(X_test)

# Filter target names and labels for classification report
unique_labels = np.unique(np.concatenate([y_test, y_pred]))
filtered_target_names = [str(le.classes_[label]) for label in unique_labels]

print("Classification report:")
print(classification_report(
    y_test, y_pred,
    labels=unique_labels,
    target_names=filtered_target_names,
    zero_division=0
))

print(f"Test set accuracy: {accuracy_score(y_test, y_pred):.4f}")
```
